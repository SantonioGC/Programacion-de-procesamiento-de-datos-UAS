```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('csv/act1.csv')
df
```

```python
media_q1 = np.mean(df['Ventas_Q1'])
desv_std_q1 = np.std(df['Ventas_Q2'])
```

```python
df['Venta_Total'] = df['Ventas_Q1'] + df['Ventas_Q2']
df['Crecimiento_%'] = ((df['Ventas_Q2'] - df['Ventas_Q1']) / df['Ventas_Q1']) * 100
```

```python
resumen = df.groupby('Sucursal').agg({
    'Venta_Total' : ['sum','mean'],
    'Vendedor' : 'count'
}).reset_index()
```
