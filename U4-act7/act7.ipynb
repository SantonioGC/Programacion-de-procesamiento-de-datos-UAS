```python
import pandas as pd
```

```python
df_productos = pd.read_csv('csv/productos.csv')
df_ventas = pd.read_csv('csv/ventas.csv')
```

```python
try:
    reporte = pd.merge(df_productos, df_ventas, on='ID_Producto', how='left', validate='one_to_many')
    print("Validación exitosa: Integridad 1:m confirmada.")
except Exception as e:
    print(f"Error en los datos: {e}")
```

```python
huérfanos = df_ventas[~df_ventas['ID_Producto'].isin(df_productos['ID_Producto'])]
```

```python
reporte['Cantidad'] = reporte['Cantidad'].fillna(0)
reporte['Total_Linea'] = reporte['Cantidad'] * reporte['Precio']
```

```python
print("\n--- REPORTE FINAL ---")
print(reporte)
print("\n--- PRODUCTOS HUÉRFANOS ---")
print(huérfanos)
```
